## 문제11

이름 : 나이썬, 김자바, 김씨, 박씨샵  
등급 :   B       B      A      C


````python
# 답 1
이름 = ("나이썬", "김자바", "김씨", "박씨샵")
등급 = ("B", "B", "A", "C")
수강생_등급 = dict(zip(이름, 등급))
print(수강생_등급)
````
